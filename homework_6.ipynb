### Получим доход по каждому методу доставки

```python
data['Shipping Type'].nunique()
```

```python
income_by_shipping_type = data[data['Order Status'] == 'Completed'].groupby(['Shipping Type'])['Total Price'].agg(['sum']).rename(columns={'sum':'Total Income'}).sort_values(by='Total Income', ascending=False)
```

```python
income_by_shipping_type
```

### Получим доход по каждому типу продукта

```python
data[data['Order Status'] == 'Completed'].groupby(['Product Type'])['Total Price'].agg(['sum']).rename(columns={'sum':'Total Income'}).sort_values(by='Total Income', ascending=False)
```

### Получим доход по дополнительным услугам за каждый месяц

```python
data['Purchase Date'] = pd.to_datetime(data['Purchase Date'])
```

```python
data.info()
```

```python
months = {
    1: 'January', 
    2: 'February', 
    3: 'March', 
    4: 'April',
    5: 'May', 
    6: 'June', 
    7: 'July', 
    8: 'August',
    9: 'September', 
    10: 'October', 
    11: 'November', 
    12: 'December'
}
```

```python
monthly_grouped = data[data['Order Status'] == 'Completed'].groupby(data['Purchase Date'].dt.month)['Total Price'].agg(['sum'])
```

```python
monthly_grouped.index = monthly_grouped.index.map(months)
```

```python
monthly_grouped
```

### Получим доход по дополнительным услугам за каждый квартал

```python
grouped_by_quarter = data[data['Order Status'] == 'Completed'].groupby(pd.Grouper(key='Purchase Date', freq='QE'))
```

```python
grouped_by_quarter['Total Price'].agg(['sum'])
```
